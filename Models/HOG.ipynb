```python
import cv2
import dlib
from matplotlib import pyplot as plt
import os
```

```python
image = cv2.imread('/content/drive/MyDrive/Dataset/train/with_mask/256-with-mask.jpg')
image = cv2.imread(image)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
detector = dlib.get_frontal_face_detector()
faces = detector(gray)
for face in faces:
    x1 = face.left() 
    y1 = face.top()
    x2 = face.right()
    y2 = face.bottom()

    cv2.rectangle(img=img, pt1=(x1, y1), pt2=(x2, y2), color=(0, 255, 0), thickness=4)

plt.imshow(img)
plt.show()
```
