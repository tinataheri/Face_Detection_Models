```python
import sys, os
import numpy as np
from math import *
import math
from PIL import Image

import cv2
import os
```

```python
def filterCoordinate(c,m):
	if c < 0:
		return 0
	elif c > m:
		return m
	else:
		return c

folds = ["01","02","03","04","05","06","07","08","09","10"]
for fold_num in folds:

  ellipse_filename = "/content/drive/MyDrive/paper_face_recognitioin/FDDB-folds/ellipsis/FDDB-fold-{f_name}-ellipseList.txt".format(f_name = fold_num)
  rect_filename = "/content/drive/MyDrive/paper_face_recognitioin/FDDB-folds/rect/FDDB-fold-{f_name}-rectList.txt".format(f_name = fold_num)

  with open(ellipse_filename) as f:
    lines = [line.rstrip('\n') for line in f]

  f = open(rect_filename,'w')
  text = ''
  i = 0
  while i < len(lines):
    img_file = '/content/drive/MyDrive/paper_face_recognitioin/FDDB-images/' + lines[i] + '.jpg'
    img = Image.open(img_file)
    w = img.size[0]
    h = img.size[1]
    num_faces = int(lines[i+1])
    text = img_file + '\n' + str(num_faces) + '\n'
    for j in range(num_faces):
      ellipse = lines[i+2+j].split()[0:5]
      a = float(ellipse[0])
      b = float(ellipse[1])
      angle = float(ellipse[2])
      centre_x = float(ellipse[3])
      centre_y = float(ellipse[4])
      
      tan_t = -(b/a)*tan(angle)
      t = atan(tan_t)
      x1 = centre_x + (a*cos(t)*cos(angle) - b*sin(t)*sin(angle))
      x2 = centre_x + (a*cos(t+pi)*cos(angle) - b*sin(t+pi)*sin(angle))
      x_max = filterCoordinate(max(x1,x2),w)
      x_min = filterCoordinate(min(x1,x2),w)
      
      if tan(angle) != 0:
        tan_t = (b/a)*(1/tan(angle))
      else:
        tan_t = (b/a)*(1/(tan(angle)+0.0001))
      t = atan(tan_t)
      y1 = centre_y + (b*sin(t)*cos(angle) + a*cos(t)*sin(angle))
      y2 = centre_y + (b*sin(t+pi)*cos(angle) + a*cos(t+pi)*sin(angle))
      y_max = filterCoordinate(max(y1,y2),h)
      y_min = filterCoordinate(min(y1,y2),h)

      face_coo = [x_min,y_min,x_max,y_max]
      face_coo = [int(float(i)) for i in face_coo]
      face_coo = ','.join([str(elem) for elem in face_coo])
      text += face_coo + '\n'

    f.write(text)
    i = i + num_faces + 2
  print(fold_num,"is done")
  f.close()
```
